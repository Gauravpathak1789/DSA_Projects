```python
import gradio as gr

class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

class StudentManagementSystem:
    def __init__(self, start=None):
        self.start = start

    def add_information(self, name, id, email, grades):
        grades = [int(g) for g in grades.split(',')]
        n = Node([name, int(id), email, grades])
        if self.start is None:
            self.start = n
        else:
            temp = self.start
            while temp.next is not None:
                temp = temp.next
            temp.next = n
        return "Information added successfully."

    def delete_information(self, id):
        id = int(id)
        if self.start is None:
            return "No records to delete."
        if self.start.data[1] == id:
            self.start = self.start.next
            return f"Record with ID {id} deleted."
        temp = self.start
        prev = None
        while temp is not None:
            if temp.data[1] == id:
                if prev:
                    prev.next = temp.next
                return f"Record with ID {id} deleted."
            prev = temp
            temp = temp.next
        return f"Record with ID {id} not found."

    def update_information(self, id, name=None, new_id=None, email=None, grades=None):
        id = int(id)
        current = self.start
        while current:
            if current.data[1] == id:
                if name:
                    current.data[0] = name
                if new_id:
                    current.data[1] = int(new_id)
                if email:
                    current.data[2] = email
                if grades:
                    current.data[3] = [int(g) for g in grades.split(',')]
                return f"Record with ID {id} updated."
            current = current.next
        return f"Record with ID {id} not found."

    def search_info(self, id):
        id = int(id)
        current = self.start
        while current:
            if current.data[1] == id:
                return current.data
            current = current.next
        return f"Record with ID {id} not found."

    def show_records(self):
        if self.start is None:
            return "No records to display."
        else:
            records = []
            temp = self.start
            while temp is not None:
                records.append(tuple(temp.data))
                temp = temp.next
            return records

# Instantiate the management system
sms = StudentManagementSystem()

def add_info(name, id, email, grades):
    return sms.add_information(name, id, email, grades)

def delete_info(id):
    return sms.delete_information(id)

def update_info(id, name=None, new_id=None, email=None, grades=None):
    return sms.update_information(id, name, new_id, email, grades)

def search_info(id):
    return sms.search_info(id)

def show_records():
    return sms.show_records()

with gr.Blocks() as app:
    gr.Markdown("# Student Management System")

    with gr.Tab("Add Information"):
        name_input = gr.Textbox(label="Name")
        id_input = gr.Textbox(label="ID")
        email_input = gr.Textbox(label="Email")
        grades_input = gr.Textbox(label="Grades (comma-separated)")
        add_button = gr.Button(label="Add")
        output = gr.Textbox(label="Output")

        add_button.click(fn=add_info, inputs=[name_input, id_input, email_input, grades_input], outputs=output)

    with gr.Tab("Delete Information"):
        delete_id_input = gr.Textbox(label="ID to delete")
        delete_button = gr.Button(label="Delete")
        delete_output = gr.Textbox(label="Output")

        delete_button.click(fn=delete_info, inputs=delete_id_input, outputs=delete_output)

    with gr.Tab("Update Information"):
        update_id_input = gr.Textbox(label="ID to update")
        update_name_input = gr.Textbox(label="New Name (optional)")
        update_new_id_input = gr.Textbox(label="New ID (optional)")
        update_email_input = gr.Textbox(label="New Email (optional)")
        update_grades_input = gr.Textbox(label="New Grades (comma-separated, optional)")
        update_button = gr.Button(label="Update")
        update_output = gr.Textbox(label="Output")

        update_button.click(fn=update_info, inputs=[update_id_input, update_name_input, update_new_id_input, update_email_input, update_grades_input], outputs=update_output)

    with gr.Tab("Search Information"):
        search_id_input = gr.Textbox(label="ID to search")
        search_button = gr.Button(label="Search")
        search_output = gr.Textbox(label="Output")

        search_button.click(fn=search_info, inputs=search_id_input, outputs=search_output)

    with gr.Tab("Show Records"):
        show_button = gr.Button(label="Show Records")
        show_output = gr.Textbox(label="Output")

        show_button.click(fn=show_records, inputs=[], outputs=show_output)

app.launch(share=True)
```
